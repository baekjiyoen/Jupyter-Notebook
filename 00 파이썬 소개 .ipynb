# 파이썬의 인기

<img src='https://www.i-programmer.info/images/stories/News/2020/jan/A/tiobe2020.jpg' width='800px' height='400px'/>



- 파이썬은 사용자 친화적이다.
    - 직관적이고, 깔끔하며, 체계적이며, 사용하기 쉬운 특징을 가진다.
- 파이썬은 인기도 많고, 수요도 높다.
    - 8만 7천명의 개발자가 참여한 스택오버플로(StackOverflow)의 개발자 설문조사에 따르면, 파이썬(Python)은 자바스크립트와, HTML/CSS, SQL 다음으로 인기있는 기술이다. 이외에도 세계에서 두 번째로 인기가 많으며, 수요는 가장 높은 언어로 뽑히기도 했다.
- 파이썬은 범용성을 지닌 언어이다.
    - 윈도우나 맥OS, 리눅스 등 어떠한 환경이든 관계없이, 파이썬은 거의 모든 개발 시나리오에서 사용될 수 있다.
    - Python은 게임, 프로그래밍 언어 개발, 프로토타입 제작, 그래픽 디자인 응용프로그램, 웹 프레임워크 등의 다양한 분야에서 핵심적인 역할을 하며, 최근에는 데이터 사이언스(data science) 및 머신러닝(ML) 분야에서 사용되는 핵심 언어 중 하나로 인기를 끌고 있다.
    - 파이썬은 단순하면서도 고급 언어이기 때문에, 머신러닝 알고리즘을 위한 빠른 프로토타입 제작과 수정 작업을 가능하게 해준다.
    
    
### 파이썬 실습할 수 있는 여러 가지 방법
- 파이썬 설치(에디터 사용), python.org
- 아나콘다 설치(주피터노트북 사용)
    - 주피터노트북 사용법
    - Code/Markdown
    - Cell 추가와 삭제
    - 프로그램 작성과 실행
- 아나콘다가 무거운 툴이기 때문에 요즘에는 미니콘다를 많이 설치함
    - 미니콘다는 아나콘다의 라이트 버전임
    - 기본적인 요구사항만 포함하고 있고, 사용하는 패키지를 그때 그때 설치해서 사용
    - https://velog.io/@ash3767/미니콘다-vs-아나콘다
    - https://docs.conda.io/en/latest/miniconda.html
- 구글 CoLab 사용(별도의 에디터나 파이썬 설치 필요하지 않음)

     - 클라우드 환경에서 주피터 노트북 기반의 파이썬 개발을 위한 환경 제공
     - 파이썬 설치하지 않고도 웹 브라우저를 이용하여 주피터 노트북에서 파이썬을 사용할 수 있음
     - numpy,pandas, tensorflow 등과 같은 데이터 분석 및 머신러닝 패키지들이 설치되어 있어서 개별적으로 설치할 필요가 없음
     - 정식명칭은 Google Colaboratory로 구글드라이버와 주피터 노트북 환경에서 협업을 통한 개발이 가능함
     - 개발을 위한 웹 사이트는 다음과 같으며 사용을 위해서는 구글계정이 필요함
 
 ### 미니콘다 설치
 1. 윈도 사용자 이름은 반드시 영문(0), 한글(x)
    * 사용자 계정 만들기 단축키: netplwiz (반드시 영문이름의 관리자 계정으로 만듬)
 2. 윈도 버전 및 시스템 종류 확인 (예: Windows10, 64bit)
    * [내 컴퓨터] 우클릭 > [속성]클릭
 3. 기본 웹브라우저 "크롬(Chrome)"으로 지정함
    * 시작-'기본 앱'-'웹 브라우저' 'Chrome'으로 지정
 4. 미니콘다 설치
    * 미니콘다 다운로드 : https://docs.conda.io/en/latest/miniconda.html
    
 ### 가상환경 
 > 프로젝트별 python버전 및 기타 패키지 버전을 다르게 사용하고자 할 때

#### 1. 가상환경 만들기
1. [시작]-[Anaconda Prompt (Miniconda3)]
2. 가상환경 만들기
> - 기본 명령: conda create -n 가상환경명 python=파이썬 버전 <br>
> - 예) conda create -n smart python=3.9

3. 가상환경 확인 : conda info --envs
4. 가상환경 활성화
> - 기본 명령: activate 가상환경명
> - 예) activate smart

5. 기본 패키지 설치
> - 쥬피터 노트북 설치: conda install jupyter

> - 기본 패키지 설치: numpy, scipy, pandas, matplotlib, seaborn, scikit-learn, pillow <br>
> 예) pip install numpy scipy pandas matplotlib seaborn scikit-learn pillow

6. 주피터 노트북 실행
- 명령 터미널에서 실행:<font size="4em" color="green"> jupyter notebook </font>

- 시작 메뉴에서 실행: <font size="4em" color="green"> [시작] > Jupyter Notebook(smart)</font>

7. 기타 명령어
> - conda 및 pip 업데이트:
        - <font size="4em" color="green">conda update conda  </font>
        - <font size="4em" color="green">pip install --upgrade pip  </font>
> - 설치된 패키지 목록 확인: <font size="4em" color="green">conda list 또는 pip list</font> <br>
> - 설치된 패키지 목록 및 버전 확인: <font size="4em" color="green">pip freeze</font> <br>
> - 특정 패키지 버전 확인 : <font size="4em" color="green"> pip show [패키지 명]</font>
> - 파이썬 버전 확인: <font size="4em" color="green"> python --version</font>
> - 패키지 업데이트: <font size="4em" color="green"> pip install [패키지명] --upgrade</font>
> - 패키지 제거 : <font size="4em" color="green"> conda uninstall [패키지명] 또는 pip uninstall [패키지명]</font>
> - 기본 base 환경으로 나가기: <font size="4em" color="green"> deactivate</font> 
>
> - 가상환경 삭제: <font size="4em" color="green"> conda env remove --n 가상환경명</font>
>

8. 패키지 목록 관리하기
> - 자주 사용하는 패키지는 목록으로 만들어 두고 사용하면 편리함 (requirements.txt)
> - 설치된 패키지 requirements.txt로 저장하기 : <font size="4em" color="green">pip freeze > requirements.txt</font>
>
> - requirements.txt 내용대로 패키지 설치: <font size="4em" color="green">pip install -r requirements.txt</font>
> - requirements.txt 내용대로 패키지 삭제: <font size="4em" color="green">pip uninstall -r requirements.txt</font>
