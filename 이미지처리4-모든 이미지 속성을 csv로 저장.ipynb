```python
import pandas as pd
from glob import glob
from PIL import Image
```

```python
#1. 파일 목록 불러오기
file_names=glob('./images/*.jpg')
# print(file_names)
#2. 각 파일의 속성 가져오기
res=[]
for file_name in file_names:
    tmp=[]
    img=Image.open(file_name)
    tmp.append(img.filename)
    tmp.append(img.size)
    tmp.append(img.width)
    tmp.append(img.height)
    tmp.append(img.mode)
    tmp.append(img.format)
    res.append(tmp)
print(res,len(res),len(file_names))
#3. 각 파일의 속성을 csv로 저장하기 (csv-> 텍스트파일, 콤마로 나누어진 데이터)
pd.DataFrame(res,columns=['파일명','크기','가로크기','세로크기','모드','포맷'])
```
