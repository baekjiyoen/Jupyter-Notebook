```python
import pandas as pd
import requests
```

```python
def serch_code(areaCode):
    url='http://apis.data.go.kr/B551011/KorService1/areaCode1'
    param={'serviceKey':'Jid5jJfVfwuc7brfbSDE6yxusvhmIOTMmmZE//+2YdeSPIAoRpSnof9qpb5OSeEpyvP8xnb6AO4V7kXVEJFPEQ==',
        'numOfRows':'1000',
        'pageNo': '1',
        'MobileOS':'ETC',
        'MobileApp':'AppTest',
        'areaCode' : areaCode,
        '_type':'json'
          }
    response = requests.get(url, params=param).json()
    res = pd.DataFrame(response['response']['body']['items']['item'])[['code','name']]
    return res
```

```python
si_df=serch_code('')
si_df
```

```python
res=[]
for si_code,si_name in si_df.values:
    tmp=serch_code(si_code)
    tmp['si_code']=si_code
    tmp['si_name']=si_name
    res.append(tmp)
```

```python
code_df=pd.concat(res,ignore_index=True)
code_df.head()
```
