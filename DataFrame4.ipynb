```python
df=pd.read_csv('./data/cctv_in_seoul.csv')
print(df.shape)
df.head()
```

```python
df['2016년 증감']=df['2016년']-df['2015년']
df
```

```python
res=[]
for i in df['2016년 증감']:
    if i>=0:
        res.append('증가')
    else:
        res.append('감소')
df['기타']=res
df.head()
```

```python
res=['증가' if i >=0 else '감소' for i in df['2016년 증감']]
df['비고1'] = res
```

```python
res=[]
for i in df['2016년']:
    if i<=292:
        tmp='하'
    elif i<=467:
        tmp='중'
    else:
        tmp='상'
    res.append(tmp)
df['구분']=res
df
```

```python
df['구분1']=pd.cut(df['2016년'],bins=3,labels=['하','중','상'])
```
