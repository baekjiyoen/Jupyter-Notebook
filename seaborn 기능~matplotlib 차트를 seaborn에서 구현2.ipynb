```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
print(df.shape)
df.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,row='time')
plt.show()
```

```python
sns.barplot(y='day',x='total_bill',data=df)
```

```python
sns.countplot(x='sex',data=df,hue='time')
```
