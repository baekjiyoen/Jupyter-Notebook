```python
import folium
import pandas as pd
```

```python
map=folium.Map(location=[36.3504567,127.3848187],zoom_start=7)
folium.Marker([36.3504567,127.3848187],
             popup='<div style="width: 50px;">대전시청</div>',
             icon=folium.Icon(icon='glyphicon glyphicon-ok')).add_to(map)
folium.Marker([35.179665,129.0747635],
             popup='<div style="width: 70px;">부산시청</div>',
             icon=folium.Icon(icon='glyphicon glyphicon-plane')).add_to(map)
folium.Marker([37.5665851,126.9782038],
             popup='서울특별시청',
             icon=folium.Icon(icon='glyphicon glyphicon-remove')).add_to(map)
map
```

```python
map=folium.Map(location=[36.3504567,127.3848187],zoom_start=7)
folium.Marker([36.3504567,127.3848187],
             popup='대전시청',
             icon=folium.Icon(icon='glyphicon glyphicon-ok')).add_to(map)

folium.CircleMarker([36.3504567,127.3848187],
                    redius=100,
                    color='#ff0000',
                    fill_colore='#fffggg',
                    popup='a').add_to(map)
map
```
