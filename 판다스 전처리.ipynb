```python
df=df[df['규모']!='전체'].reset_index(drop=True)
```

```python
# 1. 규모열에서 ㎡ 삭제
res=[]
for i in df['규모']:
    tmp=i.replace('㎡','')
    res.append(tmp)
df['규모']=res
res
```

```python
# 2. 월 -> 분기열 만들기 => 1사분기,2사분기,3사분기,4사분기 만들어가
res=[]
for i in df['월']:
    if i in [1,2,3]:
        tmp='1사분기'
    elif i in [4,5,6]:
        tmp='2사분기'
    elif i in [7,8,9]:
        tmp='3사분기'
    else:
        tmp='4사분기'
    res.append(tmp)
df['분기']=res
res
```

```python
# 3.지역-> 권역열 만들기 -> 수도권(서울,경기,인천), 지방(나머지)
res=[]
for i in df['지역']:
    if i in ['서울','인천','경기']:
        res.append=('수도권')
    else:
        res.append=('지방')
df['권역']=res
```

```python
season={
    '1사분기':'봄',
    '2사분기':'여름',
    '3사분기':'가을',
    '4사분기':'겨울'
}
df['분기'].map(season)
```
