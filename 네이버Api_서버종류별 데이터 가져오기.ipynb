```python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request
import json
import pandas as pd

#1. id,pw
client_id = "0p4Rd8xJQxVRre_lMgBa"
client_secret = "SBObTGPumD"

def naverAPI(search, sever_name='블로그',display=100,start=1):
    '''
    server_name: '블로그, 뉴스, 카페, 지식인, 쇼핑'
    '''
    #서버선택
    sever={
        '블로그':'blog',
        '뉴스':'news',
        '카페':'cafearticle',
        '지식인':'kin',
        '쇼핑':'shop'
    }
    #2. url생성
    base_url=f'https://openapi.naver.com/v1/search/{sever[sever_name]}'
    params={
        'query':search,
        'display':display,
        'start':start
    }
    enc_param=urllib.parse.urlencode(params)
    url=f'{base_url}?{enc_param}'

    # 3. url에 json자료 요청
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        respose_dict=json.loads(response_body)
        res_df=pd.DataFrame(respose_dict['items'])
    else:
        print("Error Code:" + rescode)
        return
    return res_df
# naverAPI('스타벅스',sever_name='뉴스')
res=[]
for i in range(3): #i->0,1,2
    start=i*100+1 #start ->1,101,201
    tmp=naverAPI('전세사기',sever_name='지식인',start=start)
    res.append(tmp)

res_df=pd.concat(res,ignore_index=True)
print(res_df.shape)
res_df
```
