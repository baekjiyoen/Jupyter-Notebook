## 문자열 메서드

```python
# 문자열 메서드 확인
dir(str)
```

```python
# 대문자로 변환(upper)
s = 'abCdeFg'
s.upper()
```

```python
#소문자로 변환(lower)
s = 'abCdeFg'
s.lower()
```

```python
# 문자열 바꾸기(replace)
s = 'AB-C-D-EFGH'
s.replace('-','')
```

```python
# 문자열 분리하기(spㅣit)
s='서울시 중구 명동'
print(s.split(' '))

s='2022-07-04'
print(s.split('-'))

s='홍길동///남///27'
print(s.split('///'))
```

```python
#문자열 결합(join) : 결합문자.join(리스트)
s=['2021','06','03']
'-'.join(s)
```

```python
#양쪽 공백제거 (strip)
s='   김 길동   '
s=s.strip()  # 양쪽 공백 제거
print(s)
```

```python
#양쪽 특정 문자 제거 
#문자열.strip(']') -> 양쪽 끝 ']'제거
#문자열.strip('[]') -> 양쪽 끝 ']' 또는 '[' 제거

s='[abc]'
print(s.strip(']'))
print(s.strip('[]'))
```

```python
# rstrip()은 오른쪽 공백 제거
# lstrip()은 왼쪽 공백 제거
```

```python
#문자열 위치 찾기(find 또는 index)
#rfind 또는 rindex : 오른쪽에서부터 문자열 위치 찾기
s='apple pineapple'
print(s.find('pine'))
print(s.find('a',5)) # s[5:] 슬라이싱처럼 찾음
print(s.find('ap',5,-1)) #s[5:-1] 슬라이싱
print(s.find('5')) #찾는 값이 없으면 -1을 반환
print(s.index('5')) #index함수는 find함수와 동일하지만 찾는 값이 없으면 에러 발생
```

```python
# 문자열 개수 세기(count)
s='apple pineapple'
s.count('a')
```
