```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
df=pd.read_csv('./2016년_사망교통사고.csv',encoding='cp949')
print(df.shape)
df.head()
```

```python
df.columns
```

```python
m=folium.Map(
    location=[36.3504567,127.3848187],
    zoom_start=7)

marker_cluster=MarkerCluster().add_to(m)
for i in df.index:
    loc=float(df['위도'][i]),float(df['경도'][i])
    p1=df['발생지시도'][i]
    p2=df['도로형태_대분류'][i]
    p3=int(df['사망자수'][i])
    
    popup_text=f'<div style="width: 70px;">[{p1}] 사망자수:{p3} 도로형태:{p2}</div>'
    folium.Marker(
        location=loc,
        popup=popup_text,
        icon=folium.Icon(color='red',icon='ok')
    ).add_to(marker_cluster)
m.save('./2016년_사망교통사고.html')
```
