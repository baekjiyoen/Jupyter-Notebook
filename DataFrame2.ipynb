```python
# 1. 분양가가 평균 이상인 데이터만 추출
df.describe()
df[df['분양가']>=df['분양가'].mean()]
```

```python
# 2. 서울 지역만 대상으로 분양가가 평균 이하인 데이터만 추출
# seoul_mean=df[df['지역']=='서울']['분양가'].mean()
seoul_df=df[df['지역']=='서울']
seoul_df[seoul_df['분양가']<=seoul_df['분양가'].mean()]
```

```python
# 3. 2번의 결과에서 가장 많이 나오는 규모는?
seoul_df[seoul_df['분양가']<=seoul_df['분양가'].mean()]['규모'].value_counts
```

```python
# 1. '부동산'폴더 만들어라
os.makedirs('./부동산',exist_ok=True)
# 2. '부동산'폴더에 지역별 csv로 저장하기 (서울.csv, 경기.csv)
jis=df['지역'].unique()

for i in jis:
    tmp=df[df['지역']==i].reset_index(drop=True)
    tmp.to_csv(f'./부동산/{i}.csv',encoding='cp949')
    
```
