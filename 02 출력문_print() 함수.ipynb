## 문자열 포매팅
### 문자열 포맷 코드



```python
name='홍길동'
age=15

print('내 이름은 '+name + '이고 나이는 ' + str(age)+'세 입니다.')

#방법1 : '문자열' % 값 
print('내 이름은 %s입니다.'%name) #데이터1개 출력
print('내 이름은 %s이고 나이는 %d세 입니다.' % (name,age)) #데이터 2개이상 출력

#방법2: format() 함수
print('내 이름은 {}입니다.'.format(name)) #데이터 1개 출력
print('내 이름은 {}이고 나이는 {}세 입니다'.format(name,age))# 데이터 2개 이상 출력
print('내 이름은 {1}이고 나이는{0}세입니다.{1}은 본명입니다.'.format(age,name))

#방법3: f문자열 포맷팅 , 파이썬 3.6버전부터 사용 가능
print(f'내 이름은 {name}이고 나이는 {age}세 입니다. 내년에는 {age+1}세가 됩니다.')
```

### 정렬과 자리수 지정
#### - 정렬 :
> - 왼쪽 정렬: < <br>
> - 오른쪽 정렬:> <br>
> - 가운데 정렬:^ <br>

```python
# '문자열' % 값
print('%10s' % 'hi') #10자리수에 오른쪽 정렬
print('%-10s' % 'hi') # 10자리수에 왼쪽 정렬
print('%0.4f' % 123.1234567) # 소후이하 4자리로 표현
print('%10.4f' % 123.1234567) # 소숫점을 포함한 전체길이 10, 소수이하 4자리

print('='*100)

# '문자열'.format(값)
print('{0:>10}'.format('hi')) # 10자리수에 오른쪽 정렬
print('{0:<10}'.format('hi')) # 10자리수에 왼쪽 정렬
print('{0:^10}'.format('hi')) # 10자리수에 가운데 정렬
print('{0:0.4f}'.format(123.1234567)) # 소수이하 4자리 표현
print('{0:10.4f}'.format(123.1234567)) # 소숫점을 포함한 전체길이 10, 소수이하 4자리

print('='*100)
#f'문자열'
s='hi'
n=123.1234567
print(f'{s:>10}') # 10자리수에 오른쪽 정렬
print(f'{s:<10}')  # 10자리수에 왼쪽 정렬
print(f'{s:-^10}') # 10자리수에 가운데 정렬 '-'로 채우기
print(f'{n:0.4f}') # 소수이하 4자리 표현
print(f'{n:10.4f}') # 소숫점을 포함한 전체길이 10, 소수이하 4자리
```
